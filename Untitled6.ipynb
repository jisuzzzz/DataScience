```python
import pandas as pd

file_path = 'gts-seoul.csv'
df = pd.read_csv(file_path)

# print(df.columns)
# gts-seoul.csv 파일을 불러옵니다.
#  잘 불러왔는지 확인하기 위해 다음의 작업을 수행하세요: 2011년 3월 21일 오전 6시의 습도, 풍속, 현지기압, 기온, 이슬점 온도를 추출하세요.
# 이들의 합을 계산하여 제출하세요.

target_data = '2011-03-21 06:00'
# df[df[col] == col_value] 조건식
target_data = df[df['일시'] == target_data]
total_sum = target_data['습도'] + target_data['풍속'] + target_data['현지기압'] + target_data['기온'] + target_data['이슬점 온도']
print(total_sum)
```

```python
# gts-seoul.csv 파일에서 불러온 정보를 다음의 순서에 따라 전처리 하세요.
# 전처리를 마친 후, 남은 행의 수 + 평균 습도 + 최대 풍속 + 최저 기온의 값을 계산하여 제출하세요.
# 습도, 풍속, 현지기압, 기온, 이슬점 온도 열(column) 추출하기
# nan 값을 포함하는 행 지우기 (힌트: dropna 사용)

# 열 선택하기 df['col_name']
# 열 여러개 선택하기 df[['col1', 'col2',...]]
ppd = df[['습도', '풍속', '현지기압', '기온', '이슬점 온도']]
ppd = ppd.dropna()

# len() -> 행 갯수
row_num = len(ppd)
# mean() -> 평균
avg_humidity = ppd['습도'].mean()
# max() -> 최댓값
max_wind = ppd['풍속'].max()
# min() -> 최솟값
min_temp = ppd['기온'].min()

total_sum2 = row_num + avg_humidity + max_wind + min_temp
print(total_sum2)
```

```python
# 습도와 이슬점 온도, 풍속과 이슬점 온도, 현지기압과 이슬점 온도, 기온과 이슬점 온도의 상관계수를 각각 구하고, 이들의 합을 계산하여 제출하세요.
# 상관계수 구하기 -> df['col1'].corr(df['col2']) col1 와 col2 의 상관계수
corr_humidity = ppd['습도'].corr(ppd['이슬점 온도'])
corr_wind = ppd['풍속'].corr(ppd['이슬점 온도'])
corr_pressure = ppd['현지기압'].corr(ppd['이슬점 온도'])
corr_tmep = ppd['기온'].corr(ppd['이슬점 온도'])

total_sum3 = corr_humidity + corr_wind + corr_pressure + corr_tmep
print(total_sum3)
```

```python
import torch
import torch.nn as nn
import numpy as np

# Cosine Similarity 객체 생성
cos = nn.CosineSimilarity(dim=0)

# 결과를 저장할 리스트
top3_indices_list = []
# test1_vectors -> floatTensor로 만들기
for test_vector in data['test1_vectors']:
    test_tensor = torch.FloatTensor(test_vector)
    similarities = []

    # 테스트 벡터와 모든 훈련 벡터와의 유사도 계산
    for train_vector in data['train_vectors']:
        train_tensor = torch.FloatTensor(train_vector)
        similarity = cos(train_tensor, test_tensor)
        similarities.append(similarity.item())

    # 유사도 값에 따라 상위 3개의 인덱스 찾기
    top3_indices = np.argsort(similarities)[-3:]
    top3_indices_list.append(top3_indices)

# 모든 인덱스 값을 합산
total_sum = np.sum(top3_indices_list)

print(total_sum)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = torch.from_numpy(data["train_vectors"])
y_train = torch.from_numpy(data["train_labels"])
# model = nn.Linear(입력 차원, 결과 차원)
# print(data["train_vectors"].shape[1]) => input 차원
# print(len(set(data["train_labels"]))) => output 차원
model = nn.Linear(256,3)
optim = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(10001):
  z = model(x_train)
  cost = F.cross_entropy(z, y_train)
  optim.zero_grad()
  cost.backward()
  optim.step()

  with torch.no_grad():
    if epoch % 1000 == 0:
      print(f"epoch:{epoch},cost:{cost.item()}")
```

```python
x_test = torch.from_numpy(data['test2_vectors'])
with torch.no_grad():
    test_outputs = model(x_test)
    # 예측 단계에서는 argmax 함수를 사용하여 가장 높은 확률을 갖는 클래스의 인덱스를 얻음
    predicted = torch.argmax(test_outputs, dim=1)

print(predicted)
```
